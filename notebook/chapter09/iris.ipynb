**The Iris Dataset**

三種の異なるアイリス(Setosa, Versicolour, Virginica)の花びらとがくの長さと幅のデータセット

```python
%matplotlib inline
from __future__ import division, print_function
```

```python
from keras.layers import Activation, Dense
from keras.layers.normalization import BatchNormalization
from keras.models import Sequential
from keras.utils.np_utils import to_categorical
import numpy as np
from sklearn import datasets
from sklearn.metrics import classification_report, confusion_matrix
```

```python
np.random.seed(0)
```

データを準備

```python
iris = datasets.load_iris()
X = iris.data
labels = iris.target
```

データをいくつか表示してみる

```python
random_indices = np.random.permutation(len(X))
print('X:', X[random_indices[:10]], 'labels:', labels[random_indices[:10]], sep='\n')
```

ラベルをone-hotベクトル（値がひとつだけ1で他が0のベクトル）に変換

```python
Y = to_categorical(labels)
```

変換後の正解データをいくつか表示してみる

```python
print(Y[random_indices[:10]])
```

訓練用とテスト用の添字を取得

```python
random_indices = np.random.permutation(len(X))
train_indices = random_indices[:len(X)//2]
test_indices = random_indices[len(X)//2:]
```

訓練用とテスト用にデータをそれぞれ準備

```python
X_train, X_test = X[train_indices], X[test_indices]
Y_train, Y_test = Y[train_indices], Y[test_indices]
labels_train, labels_test = labels[train_indices], labels[test_indices]
```

ロジスティック回帰

```python
model = Sequential()
model.add(Dense(3, input_dim=4))
model.add(Activation('softmax'))
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(X_train, Y_train, nb_epoch=200, batch_size=128)
```

```python
_, acc = model.evaluate(X_test, Y_test, verbose=0)
print('accuracy: {:.2f}'.format(acc))
```

```python
labels_pred = model.predict_classes(X_test, verbose=0)
print(confusion_matrix(labels_test, labels_pred))
print(classification_report(labels_test, labels_pred))
```

多層パーセプトロン

```python
model = Sequential()
model.add(Dense(16, input_dim=4))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(Dense(16))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(Dense(3))
model.add(BatchNormalization())
model.add(Activation('softmax'))
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(X_train, Y_train, nb_epoch=200, batch_size=128)
```

```python
_, acc = model.evaluate(X_test, Y_test, verbose=0)
print('accuracy: {:.2f}'.format(acc))
```

```python
labels_pred = model.predict_classes(X_test, verbose=0)
print(confusion_matrix(labels_test, labels_pred))
print(classification_report(labels_test, labels_pred))
```
